```python
new_examples = [
    {
        "query": "实现人脸识别功能，有哪个API可以支持",
        "answer": "您可以使用快驴容器增加视觉识别能力API ‘aiRecognize’ "
    },
    {
        "query": "如何验证用户的Session是否有效",
        "answer": "您可以使用API ‘checkSession’ 来检验用户当前的session"
    },
    {
        "query": "识别图片中的二维码的API",
        "answer": "MSI快驴识别图片二维码&条形码API `getCodeFromPic`支持识别图片中二维码"
    },
    {
        "query": "获取网络请求公共参数",
        "answer": "API `getCommonRequestConfig`支持获取网络请求公共参数，包括最终请求域名，headers和params公共参数"
    },
    {
        "query": "有支持计算文件Hash的API吗",
        "answer": "`getFileHash`支持获取指定文件的Hash值"
    },
    {
        "query": "怎么获取灵犀sessionId",
        "answer": "快驴标准API`getLXSessionId`支持获取灵犀的SeesionId"
    },
    {
        "query": "怎么获取PushToken",
        "answer": "快驴标准API`getPushToken`支持获取PushToken"
    },
    {
        "query": "拿到用户行走的步数，使用哪个API",
        "answer": "`getStepCount`可是获取步数，当时需要先调用计步器API"
    },
    {
        "query": "怎么来获取用户信息",
        "answer": "可以调用`getUserInfo`来获取当前用户信息"
    },
    {
        "query": "怎么判断手机是否安装了某个APP",
        "answer": "可以调用`isAppInstalled`来检测用户是否安装了APP"
    }                     
]
```

```python
from langchain.evaluation.qa import QAEvalChain
from langchain.evaluation.qa import QAGenerateChain
from langchain.chains import RetrievalQA
from langchain.chat_models import ChatOpenAI

```

```python
langchain.debug = True
llm = ChatOpenAI(temperature = 0.0)
#retriever = vectordb.as_retriever(search_type="mmr")
#retriever = vectordb.as_retriever(search_kwargs={"k": 3})
retriever = vectordb.as_retriever()
qa = RetrievalQA.from_chain_type(
    llm=llm, 
    chain_type="stuff", 
    retriever=retriever, 
    verbose=True,
    chain_type_kwargs = {
        "document_separator": "<<<<>>>>>"
    }
)

#print(qa.run(new_examples[1]["query"]))
#print(qa.run('checkSession这个API怎么使用'))

predictions = qa.apply(new_examples)

print(predictions)
```

```python
langchain.debug = True

llm = ChatOpenAI(temperature=0)

eval_chain = QAEvalChain.from_llm(llm)

graded_outputs = eval_chain.evaluate(new_examples, predictions)
```

```python
result = []
import json
for i, eg in enumerate(new_examples):
    data = {
        'title': f"Example {i}",
        'question': predictions[i]['query'],
        'real_answer': predictions[i]['answer'],
        'predicted_answer': predictions[i]['result'],
        'predicted_grade': graded_outputs[i]['text']
    }
    result.append(data)
print(result)
print('\r\n')
print(json.dumps(result))
```
